```python
import os
import pandas as pd
from sklearn.linear_model import LinearRegression

class Musteri():
    def __init__(self, tc, isim, sifre, gelir, partner_gelir, araba_degeri, ev_degeri, fatura_tutari, calisma_yili, yas):
        self.tc = tc
        self.isim = isim
        self.sifre = sifre
        self.bakiye = 0
        self.gelir = gelir
        self.partner_gelir = partner_gelir
        self.araba_degeri = araba_degeri
        self.ev_degeri = ev_degeri
        self.fatura_tutari = fatura_tutari
        self.calisma_yili = calisma_yili
        self.yas = yas



class KrediBasvurusu():
    def __init__(self):
        self.yol = '/content/drive/MyDrive/Techcareer_Essential_AI_Bootcamp/Week2/Homework_Banka_LR/bank_database.csv'
        self.veritabani = pd.read_csv(self.yol)
        self.model = LinearRegression()

        features = ['Gelir', 'Partner_Gelir', 'Araba_Degeri', 'Ev_Degeri', 'Fatura_Tutari', 'Calisma_Yili', 'Yas']
        target = ['Kredi_Talep']

        try:
          self.model.fit(self.veritabani[features], self.veritabani[target])

        except KeyError:
          print("The `Kredi_Talep` column is missing from the `bank_database.csv` file.")

    def tahmini_kredi_talep(self, tc, gelir, partner_gelir, araba_degeri, ev_degeri, fatura_tutari, calisma_yili, yas):
        yeni_musteri = pd.DataFrame({
            'Gelir': [gelir],
            'Partner_Gelir': [partner_gelir],
            'Araba_Degeri': [araba_degeri],
            'Ev_Degeri': [ev_degeri],
            'Fatura_Tutari': [fatura_tutari],
            'Calisma_Yili': [calisma_yili],
            'Yas': [yas]
        })

        tahmin_edilen_kredi = self.model.predict(yeni_musteri)
        return tahmin_edilen_kredi[0] if tahmin_edilen_kredi else 0



class Banka():
    def __init__(self):
        self.musteriler = list()
        self.kredi_basvuru = KrediBasvurusu()
        self.yol = '/content/drive/MyDrive/Techcareer_Essential_AI_Bootcamp/Week2/Homework_Banka_LR/bank_database.csv'  # Bu kısmı değiştirmeniz gerekebilir
        if not os.path.exists(self.yol):
            self.veritabani = pd.DataFrame(columns=['TC', 'Isim', 'Sifre', 'Gelir', 'Partner_Gelir', 'Araba_Degeri',
                                                    'Ev_Degeri', 'Fatura_Tutari', 'Calisma_Yili', 'Yas', 'Bakiye'])
            self.veritabani.to_csv(self.yol, index=False)
        else:
            self.veritabani = pd.read_csv(self.yol)
            for index, row in self.veritabani.iterrows():
                musteri = Musteri(row['TC'], row['Isim'], row['Sifre'], row['Gelir'], row['Partner_Gelir'],
                                   row['Araba_Degeri'], row['Ev_Degeri'], row['Fatura_Tutari'],
                                   row['Calisma_Yili'], row['Yas'])
                musteri.bakiye = row['Bakiye']
                self.musteriler.append(musteri)

    def musteri_ol(self, tc, isim, sifre, gelir, partner_gelir, araba_degeri, ev_degeri, fatura_tutari,
                   calisma_yili, yas):
        self.musteriler.append(Musteri(tc, isim, sifre, gelir, partner_gelir, araba_degeri, ev_degeri, fatura_tutari,
                                        calisma_yili, yas))
        print("Bankamıza kayıt olduğunuz için teşekkürler.")
        self.veritabani = pd.DataFrame(columns=['TC', 'Isim', 'Sifre', 'Gelir', 'Partner_Gelir', 'Araba_Degeri',
                                                'Ev_Degeri', 'Fatura_Tutari', 'Calisma_Yili', 'Yas', 'Bakiye'])
        for musteri in self.musteriler:
            self.veritabani = self.veritabani.append({
                'TC': musteri.tc,
                'Isim': musteri.isim,
                'Sifre': musteri.sifre,
                'Gelir': musteri.gelir,
                'Partner_Gelir': musteri.partner_gelir,
                'Araba_Degeri': musteri.araba_degeri,
                'Ev_Degeri': musteri.ev_degeri,
                'Fatura_Tutari': musteri.fatura_tutari,
                'Calisma_Yili': musteri.calisma_yili,
                'Yas': musteri.yas,
                'Bakiye': musteri.bakiye
            }, ignore_index=True)
        self.veritabani.to_csv(self.yol, index=False)

    def kredi_basvurusu_yap(self, tc):
      for musteri in self.musteriler:
        if tc == musteri.tc:
          tahmini_kredi = self.kredi_basvuru.tahmini_kredi_talep(
             tc, musteri.gelir, musteri.partner_gelir, musteri.araba_degeri, musteri.ev_degeri,
                    musteri.fatura_tutari, musteri.calisma_yili, musteri.yas
          )
          print(f"Tahmini Kredi Tutarınız: {tahmini_kredi}")
          return

        else:
          print("Müşteri bulunamadı.")
          input(y_menu)


banka = Banka()
menu = """
\t Bankaya Hoş Geldiniz
1) Müsteriyim
2) Müşteri olmak istiyorum
Q) Çıkış
"""

m_menu = """
\tHoş Geldiniz Sayın {}
1) Bakiye Sorgula
2) Para Yatır
3) Para Transfer Et
4) Para Çek
5) Kredi Başvurusu Yap
Q) Çıkış
"""

y_menu = "Ana menüye dönmek için enter'a basınız."

while True:
  print(menu)
  secim = input("Seciminiz: ")

  if secim == "1":
    girilen_tc = input("TC Kimlik Numaranızı Giriniz: ")
    tc_no = [str(a.tc) for a in banka.musteriler]
    #print(tc_no)
    if girilen_tc in tc_no:
      for musteri in banka.musteriler:
        print(f"Girilen TC: {girilen_tc}")
        print(f"Listedeki TC'ler: {tc_no}")
        if girilen_tc == musteri.tc:
          girilen_sifre = input("Şifrenizi giriniz: ")
          if girilen_sifre == musteri.sifre:
            while True:
              print(m_menu.format(musteri.isim))
              secim2 = input("Seçiminiz: ")
              if secim2 == "1":
                print("Bakiyeniz: ", musteri.bakiye)
                input(y_menu)
              elif secim2 == "2":
                yatirilan_tutar = int(input("Yatıracağınız miktar: "))
                onay = input("Kendi hesabınıza {} tutarında para yatırmayı onaylıyor musunuz? (E/H)".format(yatirilan_tutar))
                onay = onay.lower()
                if onay == "e":
                  musteri.bakiye += yatirilan_tutar
                  print("Paranız yatırıldı.")
                  input(y_menu)
                elif onay == "h":
                  print("İşleminiz iptal edildi.")
                  input(y_menu)
                else:
                  print("Hatalı seçim yaptınız.")
                  input(y_menu)
              elif secim2 == "3":
                hedef_tc = input("Para göndereceğiniz hesabın TC No: ")
                if hedef_tc in tc_no:
                  for musteri2 in banka.musteriler:
                    if hedef_tc == musteri2.tc:
                      transfer_tutari = int(input("Miktar: "))
                      if transfer_tutari <= musteri.bakiye:
                        onay = input("{} adlı müşterimize {} tutarında para göndermeyi onaylıyor musunuz? (E/H)".format(musteri2.isim, transfer_tutari))
                        onay = onay.lower()
                        if onay == "e":
                          musteri2.bakiye += transfer_tutari
                          musteri.bakiye -= transfer_tutari
                          print("Para başarıyla gönderildi.")
                          input(y_menu)
                        elif onay == "h":
                          print("İşlem iptal edildi.")
                          input(y_menu)
                        else:
                          print("Hatalı seçim yaptınız.")
                          input(y_menu)
                      else:
                        print("Bakiyeniz yetersiz.")
                        input(y_menu)
                else:
                  print("Yanlış TC girildi.")
                  input(y_menu)

              elif secim2 == "4":
                cekilecek_tutar = int(input("Miktar: "))
                if cekilecek_tutar <= musteri.bakiye:
                  musteri.bakiye -= cekilecek_tutar
                  print("Para çekme işlemi tamamlandı.")
                  input(y_menu)
                else:
                  print("Bakiyeniz yetersiz.")
                  input(y_menu)

              elif secim2 == "5":
                girilen_tc = input("TC Kimlik Numaranızı Giriniz: ")
                tc_no = [a.tc for a in banka.musteriler]
                if girilen_tc in tc_no:
                  for musteri in banka.musteriler:
                    if girilen_tc == musteri.tc:
                      tahmini_kredi = banka.kredi_basvurusu_yap(girilen_tc)
                      if tahmini_kredi is not None:
                        print(f"Tahmini kredi tutarınız: {tahmini_kredi}")
                      else:
                        print("Kredi Başvurusu yapılamadı")
                        input(y_menu)
                        break
              elif secim2 == "q" or secim2 == "Q":
                print("Çıkış yapılıyor.")
                input(y_menu)
                break
              else:
                input(y_menu)
        else:
          print("Girdiğiniz TC yanlış!")
          break
    else:
      print("girdiğiniz tc sistemde yok")
      break

  elif secim == "2":
        t = input("TC Kimlik Numarasını Giriniz: ")
        i = input("İsim Giriniz: ")
        s = input("Şifre Giriniz: ")
        gelir = float(input("Aylık Gelirinizi Giriniz: "))
        partner_gelir = float(input("Partnerinizin Aylık Gelirini Giriniz: "))
        araba_degeri = float(input("Aracınızın Değerini Giriniz: "))
        ev_degeri = float(input("Evinizin Değerini Giriniz: "))
        fatura_tutari = float(input("Aylık Fatura Tutarınızı Giriniz: "))
        calisma_yili = int(input("Çalışma Yılınızı Giriniz: "))
        yas = int(input("Yaşınızı Giriniz: "))
        banka.musteri_ol(t, i, s, gelir, partner_gelir, araba_degeri, ev_degeri, fatura_tutari,
                          calisma_yili, yas)
        input(y_menu)

  elif secim == "q" or secim == "Q":
        break

  else:
        input(y_menu)
```
